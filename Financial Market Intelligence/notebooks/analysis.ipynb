```python
def plot_trend(df, column, title, color, ma_window=30):

    df = df.copy()
    df['Date'] = pd.to_datetime(df['Date'])

    plt.style.use('dark_background')
    plt.figure(figsize=(14,6))

    # main line
    plt.plot(
        df['Date'],
        df[column],
        color=color,
        linewidth=2.5,
        label=f"{title}"
    )

    # markers
    plt.scatter(
        df['Date'].iloc[::50],
        df[column].iloc[::50],
        color='white',
        s=25,
        alpha=0.8,
        label='Data Points'
    )

    # moving average
    ma_col = f"MA_{ma_window}"
    df.loc[:, ma_col] = df[column].rolling(window=ma_window).mean()

    plt.plot(
        df['Date'],
        df[ma_col],
        linestyle='--',
        linewidth=2,
        label=f'{ma_window}-Day Avg'
    )

    # labels
    plt.title(title, fontsize=16, fontweight='bold')
    plt.xlabel("Date")
    plt.ylabel("Value")

    plt.grid(True, linestyle='--', linewidth=0.5, alpha=0.4)
    plt.legend(facecolor='black', edgecolor='white')

    plt.tight_layout()
    plt.show()
```

```python
#UNIVERSAL YEAR-WISE FUNCTION
def plot_yearly_avg(df, column, title, color):
    
    df = df.copy()
    df['Date'] = pd.to_datetime(df['Date'])
    df['Year'] = df['Date'].dt.year

    # yearly average
    yearly_avg = df.groupby('Year')[column].mean()

    plt.style.use('dark_background')
    fig_width = max(10, len(yearly_avg) * 0.8)
    plt.figure(figsize=(fig_width, 6))

    bars = plt.bar(
        yearly_avg.index,
        yearly_avg.values,
        color=color,
        edgecolor='white'
    )

    # value labels
    for bar in bars:
        height = bar.get_height()
        plt.text(
            bar.get_x() + bar.get_width()/2,
            height,
            f'{height:.0f}',
            ha='center',
            va='bottom',
            fontsize=9,
            color='white'
        )

    # titles
    plt.title(f"Yearly Average {title}", fontsize=16, fontweight='bold')
    plt.xlabel("Year")
    plt.ylabel("Price")

    plt.grid(axis='y', linestyle='--', linewidth=0.5, alpha=0.4)

    plt.tight_layout()
    plt.show()
```

```python
features = [
    # Metals
    ('Gold_24K_1g', 'Gold', '#FFD700'),
    ('Silver_1g', 'Silver', '#C0C0C0'),
    ('Platinum_1g', 'Platinum', '#00BFFF'),
    ('Palladium_1g', 'Palladium', '#DA70D6'),
    ('Copper_1g', 'Copper', '#FF8C00'),

    # Energy
    ('Crude_Oil_INR_per_barrel', 'Crude Oil', '#FF4500'),
    ('Brent_Oil_INR_per_barrel', 'Brent Oil', '#FF6347'),
    ('Natural_Gas_INR', 'Natural Gas', '#7CFC00'),

    # Currency
    ('USD_INR', 'USD/INR', '#00FF7F'),
    ('EUR_INR', 'EUR/INR', '#1E90FF'),
    ('GBP_INR', 'GBP/INR', '#FF1493')
]

for col, name, color in features:
    if col in df.columns:
        try:
            plot_yearly_avg(df, col, name, color)
        except Exception as e:
            print(f"Error in {name}: {e}")
```

```python
#Daily Price Change

# ensure datetime
df['Date'] = pd.to_datetime(df['Date'])
df = df.copy()

# daily change
df.loc[:, 'Gold_Change'] = df['Gold_24K_1g'].diff()

# dark theme
plt.style.use('dark_background')

plt.figure(figsize=(14,6))

# positive changes (green)
plt.plot(
    df['Date'],
    df['Gold_Change'],
    color='#00FF7F',
    linewidth=1.5,
    label='Daily Change'
)

# zero line 
plt.axhline(0, color='gray', linestyle='--', linewidth=1)

# fill positive/negative
plt.fill_between(
    df['Date'],
    df['Gold_Change'],
    where=(df['Gold_Change'] > 0),
    color='green',
    alpha=0.2
)

plt.fill_between(
    df['Date'],
    df['Gold_Change'],
    where=(df['Gold_Change'] < 0),
    color='red',
    alpha=0.2
)

# title & labels
plt.title("Daily Gold Price Change", fontsize=16, fontweight='bold', color='white')
plt.xlabel("Date", fontsize=12, color='white')
plt.ylabel("Change (INR)", fontsize=12, color='white')

# grid
plt.grid(True, linestyle='--', linewidth=0.5, color='gray', alpha=0.4)

# ticks
plt.xticks(color='white')
plt.yticks(color='white')

# legend
plt.legend(facecolor='black', edgecolor='white')

# spacing
plt.tight_layout()

plt.show()
```

```python
#Percentage Change

# ensure datetime
df['Date'] = pd.to_datetime(df['Date'])
df = df.copy()

# daily return %
df.loc[:, 'Gold_Return_%'] = df['Gold_24K_1g'].pct_change(fill_method=None) * 100

# dark theme
plt.style.use('dark_background')

plt.figure(figsize=(14,6))

# return line
plt.plot(
    df['Date'],
    df['Gold_Return_%'],
    color='#00FFFF',  # cyan
    linewidth=1.5,
    label='Daily Return (%)'
)

# zero line (important)
plt.axhline(0, color='gray', linestyle='--', linewidth=1)

# fill positive returns
plt.fill_between(
    df['Date'],
    df['Gold_Return_%'],
    where=(df['Gold_Return_%'] > 0),
    color='green',
    alpha=0.2
)

# fill negative returns
plt.fill_between(
    df['Date'],
    df['Gold_Return_%'],
    where=(df['Gold_Return_%'] < 0),
    color='red',
    alpha=0.2
)

# title & labels
plt.title("Gold Daily Return (%)", fontsize=16, fontweight='bold', color='white')
plt.xlabel("Date", fontsize=12, color='white')
plt.ylabel("Return (%)", fontsize=12, color='white')

# grid
plt.grid(True, linestyle='--', linewidth=0.5, color='gray', alpha=0.4)

# ticks
plt.xticks(color='white')
plt.yticks(color='white')

# legend
plt.legend(facecolor='black', edgecolor='white')

# spacing
plt.tight_layout()

plt.show()
```

```python
#Monthly Trend

# ensure datetime
df['Date'] = pd.to_datetime(df['Date'])
df = df.copy()

# monthly data
df['Month'] = df['Date'].dt.to_period('M')
monthly_avg = df.groupby('Month')['Gold_24K_1g'].mean()

# convert Period → datetime (IMPORTANT for plotting)
monthly_avg.index = monthly_avg.index.to_timestamp()

# dark theme
plt.style.use('dark_background')

plt.figure(figsize=(14,6))

# main line
plt.plot(
    monthly_avg.index, 
    monthly_avg.values,
    color='#FFD700',
    linewidth=2.5,
    label='Monthly Avg Price'
)

# markers
plt.scatter(
    monthly_avg.index,
    monthly_avg.values,
    color='white',
    s=15,
    alpha=0.7
)

# area fill (premium look)
plt.fill_between(
    monthly_avg.index,
    monthly_avg.values,
    color='#FFD700',
    alpha=0.1
)

# title & labels
plt.title("Monthly Gold Price Trend", fontsize=16, fontweight='bold', color='white')
plt.xlabel("Month", fontsize=12, color='white')
plt.ylabel("Price (INR)", fontsize=12, color='white')

# grid
plt.grid(True, linestyle='--', linewidth=0.5, color='gray', alpha=0.4)

# ticks
plt.xticks(color='white', rotation=45)
plt.yticks(color='white')

# legend
plt.legend(facecolor='black', edgecolor='white')

# spacing
plt.tight_layout()

plt.show()
```

```python
#Volatility

# ensure datetime
df['Date'] = pd.to_datetime(df['Date'])
df = df.copy()

# volatility (30-day std)
df.loc[:, 'Volatility'] = df['Gold_24K_1g'].rolling(window=30).std()

# dark theme
plt.style.use('dark_background')

plt.figure(figsize=(14,6))

# main volatility line
plt.plot(
    df['Date'], 
    df['Volatility'], 
    color='#00FFFF',  # cyan
    linewidth=2,
    label='Volatility (30-Day)'
)

# area fill (premium)
plt.fill_between(
    df['Date'], 
    df['Volatility'], 
    color='#00FFFF', 
    alpha=0.1
)

# highlight peaks (high volatility)
high_vol = df['Volatility'] > df['Volatility'].mean()

plt.scatter(
    df['Date'][high_vol],
    df['Volatility'][high_vol],
    color='red',
    s=15,
    label='High Volatility'
)

# title & labels
plt.title("Gold Price Volatility (30-Day)", fontsize=16, fontweight='bold', color='white')
plt.xlabel("Date", fontsize=12, color='white')
plt.ylabel("Volatility", fontsize=12, color='white')

# grid
plt.grid(True, linestyle='--', linewidth=0.5, color='gray', alpha=0.4)

# ticks
plt.xticks(color='white')
plt.yticks(color='white')

# legend
plt.legend(facecolor='black', edgecolor='white')

# spacing
plt.tight_layout()

plt.show()
```
